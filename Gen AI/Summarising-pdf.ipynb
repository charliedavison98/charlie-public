### Summaries

#### Initialise model

Used Google's default values in all cases except max_output_tokens. When reducing this then the model cut off part of the response instead of changing its output to be that length. 1 token is approximately 4 characters

```python
vertexai.init(project="PROJECT LOCATION", location="us-central1")
parameters = {
    "temperature": 0.8,
    "max_output_tokens": 512,   
    "top_p": 0.8,
    "top_k": 30
}
model = TextGenerationModel.from_pretrained("text-bison@001")
```

#### Run the model

```python
f = open("Summaries.txt", "w")
for j in range(0, len(files)):
    
    
    result = ""
    print('{} summary: '.format(files[j]))
    with pdfplumber.open('pdf/' + files[j]) as pdf:
        for i in range(0,len(pdf.pages)):
            page = pdf.pages[i]
            result += page.extract_text()
    try:
        f.write('{} summary: \n'.format(files[j]))
        response = model.predict(
        """Create a concise summary of the text in 50 words or less:
        """ + result,
        **parameters
        )
        print(f"{response.text}")
        f.write(response.text + '\n \n')  
        
    except Exception:
        print('Invalid input')
        f.write('Invalid input')
    

    print('\n')
    f.write('\n')

f.close()
    
    #"""Pick out the key words from this text. The response should be a bullet point list.
    #"""What are the top 5 words mentioned in this text and how many times are they mentioned
```
